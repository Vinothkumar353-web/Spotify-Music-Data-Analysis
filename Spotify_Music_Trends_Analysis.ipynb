```python
import numpy as np
import pandas as pd

Spotify_df = pd.read_csv("/content/archive.zip")
Spotify_df.head(10)
```

```python
Spotify_df.info()
Spotify_df.isnull().sum()

```

```python
Spotify_df.drop_duplicates(inplace=True)
Spotify_df.duplicated().sum()
```

```python

Columns_df = Spotify_df[['artist_name','album_name','popularity','year','duration_ms','instrumentalness','language']].copy()

Columns_df['duration_min'] = (Columns_df['duration_ms'] / 60000).round(0)

Columns_df.drop('duration_ms', axis=1, inplace=True)

```

```python
import matplotlib.pyplot as plt


Filtered_df = Columns_df[Columns_df['language'].str.lower() != 'unknown']


avg_popularity = (
    Filtered_df.groupby('language')['popularity']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)


plt.figure(figsize=(10, 6))
plt.scatter(avg_popularity['language'], avg_popularity['popularity'], color='teal', s=100)

plt.title("Average Popularity by Language")
plt.xlabel("Language")
plt.ylabel("Average Popularity")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



```

INFERENCES OF " POPULARITY BY LANGUAGE " CHART :

* The Spotify dataset includes songs in six different languages.

* Korean songs have the highest average popularity, indicating a strong global fanbase, while Malayalam songs have the lowest.

* Both Korean and Hindi songs have surpassed the overall average popularity, suggesting higher listener engagement.

* English and Telugu songs show similar levels of popularity, though both remain below the average threshold.

* Tamil and Malayalam rank lowest in terms of average popularity among the categorized languages.












```python

duration_popularity_avg = Columns_df.groupby('duration_min')['popularity'].mean().reset_index()


import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(duration_popularity_avg['duration_min'], duration_popularity_avg['popularity'], color='skyblue')


plt.title("Average Popularity by Song Duration")
plt.xlabel("duration(min) ")
plt.ylabel("Average Popularity")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


**Inferences from the "Popularity by Song Duration" Chart:**


* There is a significant increase in song popularity as duration increases, with the highest popularity (above 50) observed for songs longer than 65 minutes.

* Songs with durations between 25 to 30 minutes also show notable popularity, reaching nearly 40.

* Songs in the 5 to 15-minute range exhibit a gradual decline in popularity â€” from around 20 to below 5.

* Tracks with durations between 30 to 35 minutes maintain a consistently low popularity level, generally below 5.

* Overall, the chart indicates sporadic spikes in popularity at certain durations, followed by noticeable drops, highlighting inconsistent audience engagement with longer tracks.





```python
import pandas as pd
import matplotlib.pyplot as plt

Top_Artists_with_lang = (
    Spotify_df.groupby(['artist_name', 'language'])['popularity']
    .mean()
    .reset_index()
    .sort_values(by='popularity', ascending=False)
)

Top_Artists_cleaned = Top_Artists_with_lang[
    Top_Artists_with_lang['language'].str.lower() != 'unknown'
]


top10 = Top_Artists_cleaned.head(10).reset_index(drop=True)


plt.figure(figsize=(10, 6))
bars = plt.barh(top10['artist_name'], top10['popularity'], color='mediumseagreen')


for bar, lang in zip(bars, top10['language']):
    width = bar.get_width()
    plt.text(width + 0.5,
             bar.get_y() + bar.get_height() / 2,
             lang,
             va='center')

plt.xlabel("Average Popularity")
plt.title("Top 10 Artists by Average Spotify Popularity")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

```

**Inferences of Top 10 Artists By the Popularity**


* English artists dominate the chart, securing both the first and second positions with an impressive average popularity of 80.

* The list includes one Korean artist, who ranks at the third position, indicating strong international appeal.

* There are two Hindi artists and two Tamil artists, occupying the fourth and fifth positions respectively.

* Notably, Anirudh Ravichander appears as a common contributor to both Tamil songs in the chart.

* Out of the top 10 artists:

    1. 5 are English-language artists

    2. 2 are Hindi

    3. 2 are Tamil

    4. 1 is Korean

This indicates that English artists form the majority, accounting for half of the top positions.

```python

Filtered_df = Spotify_df[Spotify_df['artist_name'].str.lower() != 'unknown']


Artist_with_High_Energy = (
    Filtered_df.groupby('artist_name')['energy']
    .sum()
    .round(2)
    .sort_values(ascending=False)
)


Top10_Energetic = Artist_with_High_Energy.head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(Top10_Energetic.index, Top10_Energetic.values, color='red')
plt.title("Top 10 Energetic Music Artists")
plt.xlabel("Artist Name")
plt.ylabel("Total Energy")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



```

Inferences from the Top 10 Energetic Music Artists Chart:


* Shankar Mahadevan leads the chart with the highest total energy, indicating a consistently energetic music style across his songs.

* Madonna and Shreya Ghoshal follow in the second and third positions, showing that both international and Indian artists contribute significantly to high-energy music.

* Artists like Billy Joel, Taylor Swift, and Justin Bieber maintain strong positions in the middle of the chart, reflecting a balanced mix of energetic tracks in their discographies.

* The lower end of the top 10 includes Maroon 5, The Weeknd, SUPER JUNIOR, and *NSYNC, with slightly lower total energy but still qualifying among the most energetic.


* Overall, the chart showcases a blend of veteran artists and modern performers, emphasizing that high energy in music is a trend across generations.

```python
import matplotlib.pyplot as plt
import textwrap

Cleaned_df = Spotify_df[Spotify_df['album_name'].str.lower() != 'unknown']

Most_danceable_album = (
    Cleaned_df.groupby(['year', 'album_name'])['danceability']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

labels = [f"{year}\n{textwrap.shorten(album, width=25, placeholder='...')}" for year, album in Most_danceable_album.index]
values = Most_danceable_album.values


plt.figure(figsize=(12, 6))
plt.stem(range(len(values)), values, linefmt='orange', markerfmt='o', basefmt=" ")
plt.xticks(range(len(labels)), labels, rotation=30, ha='right')
plt.title("Top 10 Most Danceable Albums by Year")
plt.xlabel("Year - Album Name")
plt.ylabel("Total Danceability")
plt.tight_layout()
plt.show()
```

 Inferences: Top 10 Most Danceable Albums by Year


* The album "Nowhere to Hide" (2024) stands out with the highest total danceability score, clearly leading the chart.

* Two albums from 2024 dominate the top positions, indicating a strong trend in recent years for highly danceable music releases.

* Albums released in 2023 â€” such as â€œThe Summer Houseâ€ and â€œThe Boat Houseâ€ â€” also perform significantly well, occupying the mid-high ranks.

* Notably, Taylor Swift appears twice on this list with â€œfolklore: the long pond studio sessionsâ€ (2020) and â€œRed (Taylorâ€™s Version)â€ (2021), showcasing consistent danceability across different projects.

* The album â€œHybrid Theoryâ€ (2020) also ranks relatively high, indicating its lasting popularity and rhythmic appeal despite being a remastered release.

* From the earlier years, â€œPlayback: Justu Kuthuâ€ (2015) is the only album from the previous decade, still managing to secure a position in the top 10, reflecting long-term listener engagement.

* Overall, the plot reflects that recent albums (post-2020) dominate the danceability metric, which may imply modern production trends favor more rhythmic and upbeat compositions.
